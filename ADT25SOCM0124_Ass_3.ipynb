```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print('Libraries imported.')
```

```python
url = '/content/Titanic-Dataset.csv'
df = pd.read_csv(url)

features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

df[features] = df[features].apply(pd.to_numeric, errors='coerce')
df.head(5)
```

```python
X = df[features]
y = df['Survived']

X.fillna(X.mean(), inplace=True)
```

```python
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
X = df[features].copy() # Use .copy() to avoid SettingWithCopyWarning
y = df['Survived']

X['Age'].fillna(X['Age'].mean(), inplace=True)

print(X.shape, y.shape)
print(X.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)

print('Train size:', X_train.shape)
print('Test size:', X_test.shape)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
print('Model trained.')
```

```python
y_pred = knn.predict(X_test)
list(y_pred)[:10]
```

```python
results = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred
})
results['Status'] = results.apply(lambda r: 'Correct' if r.Actual == r.Predicted else 'Wrong', axis=1)
results.head(20)
```

```python
print('Accuracy:', accuracy_score(y_test, y_pred))
print('\nClassification Report:')
print(classification_report(y_test, y_pred))
print('\nConfusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```

Conclusion

This experiment aimed to classify Titanic passengers as 'Survived' or 'Not Survived' using the k-Nearest Neighbors (kNN) algorithm. We followed these steps:

Data Loading and Preprocessing: The Titanic dataset was loaded, and relevant numerical features ('Pclass', 'Age', 'SibSp', 'Parch', 'Fare') were selected. Crucially, missing values in the 'Age' column were imputed with the mean of the 'Age' column to ensure all features were numerical and complete, as required by the kNN algorithm.

Data Splitting: The dataset was split into training (70%) and testing (30%) sets to evaluate the model's performance on unseen data. random_state=42 was used to ensure reproducibility.

Model Training: A kNN classifier was initialized with n_neighbors=3 and trained on the preprocessed training data.

Prediction: The trained model made predictions on the test set.

Evaluation: The model's performance was evaluated using several metrics:

Accuracy: The overall accuracy of the model was approximately 64.55%. This indicates that the model correctly predicted the survival status for about 64.55% of the passengers in the test set.
Classification Report: This report provided a detailed breakdown of precision, recall, and f1-score for each class (0 = Not Survived, 1 = Survived).
For class 0 (Not Survived), the model showed a precision of 0.67 and a recall of 0.79, resulting in an f1-score of 0.72. This means it was relatively good at identifying non-survivors.
For class 1 (Survived), the model had a precision of 0.60 and a recall of 0.44, with an f1-score of 0.51. This suggests that the model struggled more with correctly identifying survivors, often missing actual survivors (lower recall).
Confusion Matrix: The confusion matrix [[124 33], [62 49]] further illustrated the model's performance:
True Negatives (TN): 124 (Correctly predicted as Not Survived)
False Positives (FP): 33 (Incorrectly predicted as Survived when they did not)
False Negatives (FN): 62 (Incorrectly predicted as Not Survived when they did)
True Positives (TP): 49 (Correctly predicted as Survived)
Conclusion: The kNN model, with n_neighbors=3, achieved a moderate accuracy of 64.55%. While it performed reasonably well in identifying passengers who did not survive (higher recall for class 0), its ability to correctly identify survivors was weaker (lower recall for class 1). This imbalance suggests that the model might be biased towards predicting the majority class (non-survivors) or that the chosen features and k value are not optimally capturing the patterns for survival. Further improvements could involve exploring different values for n_neighbors, feature engineering, or trying other classification algorithms.
